```python
# Withdrawing Exercise (Corrected & Clean Code)  ##perfrom-by-aditya-soni

balance = 85000
pin = 1978
attempts = 0

while attempts < 3:
    user_pin = input("Enter your PIN: ")

    # Check PIN is numeric
    if not user_pin.isdigit():
        print("PIN must be a number!")
        attempts += 1
        print(f"Attempts left: {3 - attempts}\n")
        continue

    user_pin = int(user_pin)

    # Match PIN
    if user_pin == pin:
        print("\nPIN verified successfully!")
        print("Select Operation:")
        print("1. Amount Deposit")
        print("2. Amount Withdrawal")

        choice = input("Enter your operation (1 or 2): ")

        if choice not in ["1", "2"]:
            print("Invalid choice!")
            break

        try:
            amount = int(input("Enter amount: "))
        except ValueError:
            print("Invalid input! Please enter amount in numbers only.")
            break

        # Deposit
        if choice == "1":
            balance += amount
            print(f"Amount Deposited Successfully! New Balance: {balance}")

        # Withdrawal
        elif choice == "2":
            if amount > balance:
                print("Insufficient balance!")
            else:
                balance -= amount
                print(f"Amount Withdrawn Successfully! Remaining Balance: {balance}")

        break

    else:
        print("Incorrect PIN!")
        attempts += 1
        print(f"Attempts left: {3 - attempts}\n")

# If 3 wrong attempts
if attempts == 3:
    print("Your card has been blocked due to 3 incorrect attempts.")
```
